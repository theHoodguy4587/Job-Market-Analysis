```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../data/preprocessed/wwr_jobs_cleaned.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['job_category'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='job_category', order=df['job_category'].value_counts().index, palette='viridis')
plt.title('Job Category Distribution')
plt.xlabel('Number of Jobs')
plt.ylabel('Job Category')
plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['company'].value_counts().index[:10], df['company'].value_counts()[:10], color='skyblue')
plt.title('Top 10 Companies by Job Count')
plt.xlabel('Company')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45, ha='right')
```
